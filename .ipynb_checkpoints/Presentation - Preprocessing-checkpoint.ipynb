```python
numeric_cols = ["Age", "Credit amount", "Duration"]
```

```python
stats = {"Stat": []}
for col in numeric_cols:
    stats[col] = []
```

```python
def add_stat(df, stats, stat_name, stat_func):
    stats["Stat"].append(stat_name)
    for col in numeric_cols:
        stats[col].append(stat_func(df[col].values))
    return stats
    
```

```python
stat_functions = {
    "Min": np.min,
    "Max":np.max,
    "Mean":np.mean,
    "Std": np.std,
}
    
```

```python
for stat_function in stat_functions:
    stats = add_stat(data, stats, stat_function, stat_functions[stat_function])
```

```python
stats_df = pd.DataFrame(stats)
```

```python
print(stats_df.to_string(index=False, float_format="%.2f"))
```

```python
job_index2word = {
    0: "unskilled and non-resident", 
    1: "unskilled and resident", 
    2: "skilled", 
    3: "highly skilled"
}
def assign_job_type(col):
    return job_index2word[col]
```

```python
data.Job = data.Job.apply(assign_job_type)
```
